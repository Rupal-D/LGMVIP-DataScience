```python
# importing required libraries

# for image processing
import cv2 as cv
from google.colab.patches import cv2_imshow    

# image editing
from skimage import io
from PIL import Image
```

```python
# Reading an image from URL

url = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQbI84or72NzUUiUpuzcCcAmtdVAcWg6rTceCiA19Lii8dK8OSvVno4m_oXyZgXAYIc-OA&usqp=CAU"
```

```python
# reading an image and displaying it

image = io.imread(url)
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
cv2_imshow(image)
```

```python
# obtaining grey scale image

gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
# inverting the grey scale image

inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

```python
# blur image

blurred = cv.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
# Inverting the blur image

inverted_blurred = 255 - blurred
pencil_sketch = cv.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
```

```python
# display as subplots

cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
